# Data Cleaning

### Checking for the NULL Values

```python
print(data.isnull().sum())
```

### Filling the Missing or NULL Values

```python
data['Gender'] = data['Gender'].fillna('Male')
data['Married'] = data['Married'].fillna('Yes')
data['Dependents'] = data['Dependents'].fillna('0')
data['Self_Employed'] = data['Self_Employed'].fillna('No')
data['Credit_History'] = data['Credit_History'].fillna('1.0')
```

```python
# data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mode()[0],inplace = True)
# data['LoanAmount'].fillna(data['LoanAmount'].median(),inplace = True)
```

```python
data.fillna({
    'LoanAmount': data['LoanAmount'].median(),
    'Loan_Amount_Term': data['Loan_Amount_Term'].mode()[0]
}, inplace=True)
```

### Rechecking the NULL Values

```python
print(data.isnull().sum())
```

# Data Visualization & Analysis

### Loan Status Count

```python
sns.countplot(x='Loan_Status', data=data)
plt.title("Loan Approval Status")
plt.show()
```

### Loan Status by Education Level

```python
sns.countplot(x='Education', hue='Loan_Status', data=data)
plt.title("Loan Status by Education Level")
plt.xlabel("Education")
plt.ylabel("Count")
plt.show()
```

### Loan Status by Property Area

```python
sns.countplot(x='Property_Area', hue='Loan_Status', data=data)
plt.title("Loan Status by Property Area")
plt.xlabel("Property Area")
plt.ylabel("Count")
plt.show()
```

### EMI (Estimated Monthly Installment)

```python
data['Loan_Amount_Term_Years'] = data['Loan_Amount_Term'] / 12
data['EMI'] = data['LoanAmount'] / data['Loan_Amount_Term_Years']
```

```python
data['Total_Income'] = data['ApplicantIncome'] + data['CoapplicantIncome']
```

```python
data['Income_Loan_Ratio'] = data['Total_Income'] / data['LoanAmount']
```

### Loan Amount Distribution

```python
sns.histplot(data['LoanAmount'].dropna(), bins=30, kde=True)
plt.title("Loan Amount Distribution")
plt.xlabel("Loan Amount")
plt.ylabel("Frequency")
plt.show()
```

### Applicant Income vs. Loan Amount (Scatter Plot)

```python
sns.scatterplot(x='ApplicantIncome', y='LoanAmount', hue='Loan_Status', data=data)
plt.title("Applicant Income vs Loan Amount")
plt.xlabel("Applicant Income")
plt.ylabel("Loan Amount")
plt.show()
```

### Loan Status by Gender (Percentage)

```python
gender_status = pd.crosstab(data['Gender'], data['Loan_Status'], normalize='index') * 100
gender_status.plot(kind='bar', stacked=True, colormap='Set2')

plt.title("Loan Status by Gender (Percentage)")
plt.ylabel("Percentage")
plt.xlabel("Gender")
plt.legend(title="Loan Status", loc='upper right')
plt.xticks(rotation=0)
plt.show()
```

```python
sns.countplot(x='Gender', hue='Loan_Status', data=data, palette='Set2', order=data['Gender'].dropna().unique())
plt.title("Loan Status by Gender")
plt.show()
```

```python
sns.countplot(x='Education', hue='Loan_Status', data=data, palette='Set1')
plt.title("Loan Status by Education")
plt.show()
```

```python
property_counts = data['Property_Area'].value_counts()
property_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3'))
plt.title("Distribution of Property Area")
plt.ylabel('')
plt.show()
```

```python
sns.countplot(x='Credit_History', hue='Loan_Status', data=data, palette='coolwarm')
plt.title("Loan Status by Credit History")
plt.show()
```

```python
numeric_data = data.select_dtypes(include=['number'])

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='YlGnBu', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.histplot(data=data, x='Loan_Amount_Term', hue='Loan_Status', bins=20, kde=True)
plt.title("Loan Amount Term Distribution by Loan Status")
plt.show()
```

```python
sns.countplot(x='Married', hue='Loan_Status', data=data, palette='magma')
plt.title("Loan Status by Marital Status")
plt.show()
```

```python
sns.pairplot(data[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Total_Income', 'Loan_Status']], hue='Loan_Status')
plt.show()
```
